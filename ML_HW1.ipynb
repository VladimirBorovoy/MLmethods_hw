```python
# Задание 1.4 - Сформируйте DataFrame из исходного для региона в соответствии с Вашим вариантом: Americas

countries_by_region_df = pd.read_excel("data/regions.xlsx") # загрузка таблицы с регионами

# создание датафрейма со странами Америки
american_region_df = countries_by_region_df.loc[countries_by_region_df['Region'] == 'AME'] 

# Поиск стран Америки из исходного датафрейма
american_countries_df = df[df["Code"].isin(american_region_df["Code"])] 
```

```python
# Задание 1.6 - Постройте графики индекса WGI за 1996-2022 для стран своего региона (estimate).

# для удобства переименуем необходимые столбцы
american_countries_df = american_countries_df.rename(columns = {'Country/Territory' : 'Country',
                                                               'Estimate' : 'Estimate.1996',
                                                               'Estimate.1' : 'Estimate.1998',
                                                               'Estimate.2' : 'Estimate.2000',
                                                               'Estimate.3' : 'Estimate.2002',
                                                               'Estimate.4' : 'Estimate.2003',
                                                               'Estimate.5' : 'Estimate.2004',
                                                               'Estimate.6' : 'Estimate.2005',
                                                               'Estimate.7' : 'Estimate.2006',
                                                               'Estimate.8' : 'Estimate.2007',
                                                               'Estimate.9' : 'Estimate.2008',
                                                               'Estimate.10' : 'Estimate.2009',
                                                               'Estimate.11' : 'Estimate.2010',
                                                               'Estimate.12' : 'Estimate.2011',
                                                               'Estimate.13' : 'Estimate.2012',
                                                               'Estimate.14' : 'Estimate.2013',
                                                               'Estimate.15' : 'Estimate.2014',
                                                               'Estimate.16' : 'Estimate.2015',
                                                               'Estimate.17' : 'Estimate.2016',
                                                               'Estimate.18' : 'Estimate.2017',
                                                               'Estimate.19' : 'Estimate.2018',
                                                               'Estimate.20' : 'Estimate.2019',
                                                               'Estimate.21' : 'Estimate.2020',
                                                               'Estimate.22' : 'Estimate.2021',
                                                               'Estimate.23' : 'Estimate.2022',})
```

```python
# Сформируем новый датафрейм. Строки - значения показателя Estimate для каждого года, столбцы - страны Америки.

# поиск столбцов Estimate
estimate_columns = [col for col in american_countries_df.columns if col.startswith('Estimate')]

# преобраузем столбцы Estimate в строки
melted_df = pd.melt(american_countries_df, id_vars=['Country'], value_vars=estimate_columns, var_name='Estimate')

# Используем метод pivot для создания нового DataFrame
new_df = melted_df.pivot(index='Estimate', columns='Country', values='value')

new_df
```

```python
plt.figure(figsize=(30,20)) # Устанавливаем размер графика
plt.title('Показатель "Estimate" для стран c 1996 по 2022 год',fontsize=30) # Верхняя подпись к графику

g = sns.lineplot(data=new_df,
                linewidth = 5, # жирность линий
                 markers=True, markersize=15, markerfacecolor='red', # настройка маркеров
                dashes=False) # отключение пунктира

g.get_legend().remove() # отключаем легенду
plt.grid() # включаем сетку
plt.tick_params(axis='y', which='major', labelsize=40) # настройка оси У
plt.xticks(rotation='vertical') # поворачиваем значения оси Х
plt.tick_params(axis='x', which='major', labelsize=30) # настройка оси Х
```

```python
# Задание 1.7 - Найдите страны с наибольшим и наименьшим значением WGI Вашего варианта региона за 2022 год (estimate)

# Находим страны с наибольшим и наименьшим значениями Estimate за 2022 год
max_estimate_country = american_countries_df['Estimate.2022'].idxmax()
min_estimate_country = american_countries_df['Estimate.2022'].idxmin()

# Получаем значения этих стран
max_estimate_value = american_countries_df.loc[max_estimate_country, 'Estimate.2022']
min_estimate_value = american_countries_df.loc[min_estimate_country, 'Estimate.2022']

print("Страна с наибольшим значением Estimate за 2022 год:", american_countries_df.loc[max_estimate_country, "Country"])
print("Значение:", max_estimate_value)

print("Страна с наименьшим значением Estimate за 2022 год:", american_countries_df.loc[min_estimate_country, "Country"])
print("Значение:", min_estimate_value)
```
