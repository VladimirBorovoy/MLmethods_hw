```python
# Сформируем новый датафрейм. Строки - значения показателя Estimate для каждого года, столбцы - страны Америки.

# поиск столбцов Estimate
estimate_columns = [col for col in american_countries_df.columns if col.startswith('Estimate')]

# преобраузем столбцы Estimate в строки
melted_df = pd.melt(american_countries_df, id_vars=['Country'], value_vars=estimate_columns, var_name='Estimate')

# Используем метод pivot для создания нового DataFrame
new_df = melted_df.pivot(index='Estimate', columns='Country', values='value')

new_df
```

```python
# Задание 1.11 Определите, как изменилось значение показателя rank с 1996 по 2022 (rank)

plt.figure(figsize=(20, 30))

# Для каждой страны строим график изменения rank
for index, row in american_countries_df.iterrows():
    country = row['Country']
    rank_1996 = row['Rank']
    rank_2022 = row['Rank.22']
    
    plt.plot(['1996', '2022'], [rank_1996, rank_2022], marker='o', label=country)

plt.xlabel('Год')
plt.ylabel('Rank')
plt.title('Изменение показателя rnak 1996-2022')

plt.legend(bbox_to_anchor=( 1.05 , 1 ), loc='upper left', borderaxespad= 0 ) # выносим легенду за график
plt.grid(True)
plt.show()
```

```python
# Задание 1.12 - Выведите таблицу для Вашего варианта (WGI - rank)

# Расчеты

# Находим страны с наибольшим и наименьшим значениями rank за 1996 год
max_rank_country_1996 = american_countries_df['Rank'].idxmax()
min_rank_country_1996 = american_countries_df['Rank'].idxmin()

# Получаем значения этих стран
max_rank_value_1996 = american_countries_df.loc[max_rank_country, 'Rank']
min_rank_value_1996 = american_countries_df.loc[min_rank_country, 'Rank']

print("Страна с наибольшим значением rank за 1996 год:", american_countries_df.loc[max_rank_country_1996, "Country"])
print("Значение:", max_rank_value_1996)

print("Страна с наименьшим значением rank за 1996 год:", american_countries_df.loc[min_rank_country_1996, "Country"])
print("Значение:", min_rank_value_1996)


# Находим страны с наибольшим и наименьшим значениями rank за 2022 год
max_rank_country_2022 = american_countries_df['Rank.22'].idxmax()
min_rank_country_2022 = american_countries_df['Rank.22'].idxmin()

# Получаем значения этих стран
max_rank_value_2022 = american_countries_df.loc[max_rank_country, 'Rank.22']
min_rank_value_2022 = american_countries_df.loc[min_rank_country, 'Rank.22']

print("Страна с наибольшим значением rank за 2022 год:", american_countries_df.loc[max_rank_country_2022, "Country"])
print("Значение:", max_rank_value_2022)

print("Страна с наименьшим значением rank за 2022 год:", american_countries_df.loc[min_rank_country_2022, "Country"])
print("Значение:", min_rank_value_2022)
```

```python

# Строки и столбцы для таблицы
rows = ["mean_2022", "max_2022", "min_2022", "Russia_2022"]
cols = ["Регион", "Страна", "WGI 1996", "WGI 2022", "Изменение"]

table = pd.DataFrame(index = rows, columns = cols) # Таблица имеет тип датафрейм

# Заполняем регион
table.loc["mean_2022", "Регион"] = "AME"
table.loc["max_2022", "Регион"] = "AME"
table.loc["min_2022", "Регион"] = "AME"
table.loc["Russia_2022", "Регион"] = "ECA"

# Заполняем страну
table.loc["mean_2022", "Страна"] = ""
table.loc["max_2022", "Страна"] = american_countries_df.loc[max_rank_country_2022, "Country"]
table.loc["min_2022", "Страна"] = american_countries_df.loc[min_rank_country_2022, "Country"]
table.loc["Russia_2022", "Страна"] = "Russian Federation"

# Заполняем rank за 1996 год
table.loc["mean_2022", "WGI 1996"] = american_countries_df["Rank"]. mean()
table.loc["max_2022", "WGI 1996"] = max_rank_value_1996
table.loc["min_2022", "WGI 1996"] = min_rank_value_1996
table.loc["Russia_2022", "WGI 1996"] = df.loc[df["Country/Territory"] == "Russian Federation", "Rank"].values[0]

# Заполняем rank за 2022 год
table.loc["mean_2022", "WGI 2022"] = american_countries_df["Rank.22"]. mean()
table.loc["max_2022", "WGI 2022"] = max_rank_value_2022
table.loc["min_2022", "WGI 2022"] = min_rank_value_2022
table.loc["Russia_2022", "WGI 2022"] = df.loc[df["Country/Territory"] == "Russian Federation", "Rank.22"].values[0]

# Заполняем изменение rank
table.loc["mean_2022", "Изменение"] = american_countries_df["Rank.22"]. mean() - american_countries_df["Rank"]. mean()
table.loc["max_2022", "Изменение"] = max_rank_value_2022 - max_rank_value_1996
table.loc["min_2022", "Изменение"] = min_rank_value_2022 - min_rank_value_1996
table.loc["Russia_2022", "Изменение"] = df.loc[df["Country/Territory"] == "Russian Federation", "Rank.22"].values[0] - df.loc[df["Country/Territory"] == "Russian Federation", "Rank"].values[0]

# Вывод таблицы
table
```

```python
# Задание 1.13 Отобразите диаграмму размаха (boxplot) индекса WGI за 2022 для всех стран и для каждого региона в отдельности (на одном графике) (estimate)

countries_by_region_df = pd.read_excel("data/regions.xlsx") # загрузка таблицы с регионами
regions: list = countries_by_region_df["Region"].unique()

def df_by_region(code: str):
    temp_df = countries_by_region_df.loc[countries_by_region_df['Region'] == code]
    temp_df = df[df["Code"].isin(temp_df["Code"])]
    return temp_df

#regions_dict: dict = {}
#for region in regions:
#    regions_dict["{0}_countries_df".format(region)] = df_by_region(region)

AP_countries_df = df_by_region("AP")
ECA_countries_df = df_by_region("ECA")
MENA_countries_df = df_by_region("MENA")
SSA_countries_df = df_by_region("SSA")
AME_countries_df = df_by_region("AME")
WEEU_countries_df = df_by_region("WE/EU")

AP_countries_df = AP_countries_df.loc[:, AP_countries_df.columns.intersection(['Country/Territory','Estimate.22'])]
ECA_countries_df = ECA_countries_df.loc[:, ECA_countries_df.columns.intersection(['Country/Territory','Estimate.22'])]
MENA_countries_df = MENA_countries_df.loc[:, MENA_countries_df.columns.intersection(['Country/Territory','Estimate.22'])]
SSA_countries_df = SSA_countries_df.loc[:, SSA_countries_df.columns.intersection(['Country/Territory','Estimate.22'])]
AME_countries_df = AME_countries_df.loc[:, AME_countries_df.columns.intersection(['Country/Territory','Estimate.22'])]
WEEU_countries_df = WEEU_countries_df.loc[:, WEEU_countries_df.columns.intersection(['Country/Territory','Estimate.22'])]

AP_countries_df['Region'] = "AP"
ECA_countries_df['Region'] = "ECA"
MENA_countries_df['Region'] = "MENA"
SSA_countries_df['Region'] = "SSA"
AME_countries_df['Region'] = "AME"
WEEU_countries_df['Region'] = "WEEU"


concatenated_df = pd.concat([AP_countries_df, ECA_countries_df, MENA_countries_df, SSA_countries_df, AME_countries_df, WEEU_countries_df])

all_df = concatenated_df.copy()
all_df['Region'] = 'all'

concatenated_df = pd.concat([concatenated_df, all_df])

sns.boxplot(data=concatenated_df, x="Region", y="Estimate.22")
```
